```python
import os
import pickle
import mediapipe as mp
import cv2

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles

hands = mp_hands.Hands(static_image_mode=True, min_detection_confidence=0.3)

DATA_DIR = './dataset'
EXPECTED_LENGTH = 84  # 21 puntos clave * 2 coordenadas * 2 manos = 84

data = []
labels = []

for dir_ in os.listdir(DATA_DIR):
    for img_path in os.listdir(os.path.join(DATA_DIR, dir_)):
        data_aux = []
        x_ = []
        y_ = []

        img = cv2.imread(os.path.join(DATA_DIR, dir_, img_path))
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        results = hands.process(img_rgb)

        if results.multi_hand_landmarks:
            hand_data = []  # Para almacenar datos de cada mano detectada

            for hand_landmarks in results.multi_hand_landmarks:
                single_hand_data = []
                x_ = []
                y_ = []

                for i in range(len(hand_landmarks.landmark)):
                    x = hand_landmarks.landmark[i].x
                    y = hand_landmarks.landmark[i].y
                    x_.append(x)
                    y_.append(y)

                for i in range(len(hand_landmarks.landmark)):
                    x = hand_landmarks.landmark[i].x
                    y = hand_landmarks.landmark[i].y
                    single_hand_data.append(x - min(x_))
                    single_hand_data.append(y - min(y_))

                hand_data.append(single_hand_data)

            # Asegurarse de que haya datos de 2 manos
            if len(hand_data) == 1:  # Solo una mano detectada
                hand_data.append([0] * 42)  # Rellenar con ceros

            # Combinar los datos de ambas manos
            for hand in hand_data:
                data_aux.extend(hand)

            # Verificar longitud antes de agregar
            if len(data_aux) == EXPECTED_LENGTH:
                data.append(data_aux)
                labels.append(dir_)

# Guardar los datos y etiquetas en un archivo pickle
with open('./data/data_palabras1.pickle', 'wb') as f:
    pickle.dump({'data': data, 'labels': labels}, f)
```
